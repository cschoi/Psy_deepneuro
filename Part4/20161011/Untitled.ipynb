## 1. Characterizing a Distribution

### 1) Distribution Center

#### mean (평균)

```python
import numpy as np
x = np.arange(10)
np.mean(x)
```

```python
x_with_Nan = np.hstack((x, np.nan))
x_with_Nan
```

```python
np.mean(x_with_Nan)
```

```python
np.nanmean(x_with_Nan)
```

#### median (중앙값)

```python
np.median(x)
```

#### mode (최빈값)

```python
from scipy import stats
data = [1, 3, 4, 4, 7]
stats.mode(data)
```

#### Geometric mean (기하평균)

```python
x = np.arange(1, 101)
x
```

```python
stats.gmean(x)
```

## 2. Discrete Distributions

### 1) Bernoulli Distribution 

```python
from scipy import stats
p = 0.4
bernoulliDist = stats.bernoulli(p)
```

```python
bernoulliDist.__dict__
```

```python
p_tails = bernoulliDist.pmf(0)
p_tails
```

```python
p_heads = bernoulliDist.pmf(1)
p_heads
```

```python
trials = bernoulliDist.rvs(10)
trials
```
